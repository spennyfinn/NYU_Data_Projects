1. What is the most frequent word in "The Odyssey" and "The Epic of Gilgamesh..."?


Opening "The Odyssey" file

```python
with open('data/Homer.txt.utf-8', 'r') as file:
    content= file.read()
```

Cleaning the text so that everything is in lowercase and all punctuation is removed

```python
content = content.lower()
punctuation = ",.!?;:''\"()[]{}<>-_`~@#$%^&*+=|/\\"
cleaned_text = ''.join(word for word in content if word not in punctuation)
```

1. Making a dictionary to hold the key pair values.
2. Splitting the cleaned text into a list to iterate over
3. Iterate over the list:
4. if the word is not in the dictionary, add it to the dictionary and set the value equal to 1
5. Otherwise, increment the value associated with the key by one

```python
d = {}
li = cleaned_text.split()
for word in li:
    if word not in d:
        d[word]=1
    else:
        d[word]+=1

```
