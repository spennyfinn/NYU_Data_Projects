# 3b. Columns to be changed
* CRASH DATE -- datetime 
* CRASH TIME -- datetime
* ZIP CODE -- string

CLEAN UP
* ON STREET NAME -- making data unform with .upper(), .strip():
* BOROUGH -- making data uniform by using .upper() and. strip. Using fillna("OTHER") for my graph 

```python
# transforming CRASH DATE to datetime dtype

#importing datetime
from datetime import datetime
#converting the crash date col to datetime
df["CRASH DATE"] = pd.to_datetime(df["CRASH DATE"], format="%m/%d/%y")
#print the first 5 rows of the col
df["CRASH DATE"].head()
```

```python
#transforming CRASH TIME to datetime
df["CRASH TIME"] = pd.to_datetime(df["CRASH TIME"], format="%H:%M:%S").dt.time
df["CRASH TIME"].head()
```

```python
#transforming ZIPCODE to string from float

df["ZIP CODE"] = df["ZIP CODE"].astype(str)
print(df.columns)
```

```python
#Convert ON STREET NAME vals to a string, uppercase, and strip of white spaces
df['ON STREET NAME'] = df['ON STREET NAME'].str.upper().str.strip()
```

```python
#Clean the BOUROUGH data

#ensures that the data is uniform
df["BOROUGH"] = df["BOROUGH"].str.upper().str.strip()
#makes all the NaN data "NA"
df["BOROUGH"] = df["BOROUGH"].fillna("OTHER")
#Printing the data to see the unique vals
df["BOROUGH"].value_counts()
```

## 8.What are the distributions of accidents based on the geo location (latitude & longitude)?
* imported numpy
* make a list comp to deal with the lat conditions and long conditions to filter the table
* filter the table based on the conditions
* make a colormap dictionary
* map the colors to the boroughs in the DF
* make sure the LAT/LONG cols have the same num of values as COLOR by removing null vals
* make the scatterplot
* set the labels and title

```python
import numpy as np
#remove null vals and ensure that data remains in the right bounds for lat and long
lat_condition = (
    df["LATITUDE"].notna() & 
    (df["LATITUDE"] != 0) & 
    (df["LATITUDE"] >= -90) & 
    (df["LATITUDE"] <= 90)
)

long_condition = (
    df["LONGITUDE"].notna() & 
    (df["LONGITUDE"] >= -180) & 
    (df["LONGITUDE"] <= 180)
)

#Filter the DataFrame based on the conditions
df = df[lat_condition & long_condition]


color_map = {
    'MANHATTAN': 'skyblue',
    'BROOKLYN': 'magenta',
    'QUEENS': 'green',
    'BRONX': 'red',
    'STATEN ISLAND': 'maroon'
}

#Map the colors to the boroughs in the DataFrame
df['COLOR'] = df['BOROUGH'].map(color_map)

#remove null vals in color
df=df[df["COLOR"].notna()]



#Figure:
plt.figure(figsize=(10, 6))

plt.scatter(x = df["LATITUDE"],
            y=df["LONGITUDE"], 
            alpha=.08, 
            c = df["COLOR"]
            )

plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.title("Geographical Distribution of Accidents in NYC Color Mapped by Borough")
plt.grid(True)

plt.show()

```

## 9. Shows the covariance between each pair of columns
* I altered the date time data that I transformed earlier in the notebook (code block 7 and 8) since I needed data that could be transformed into a float. Therefore I used an apply() to convert the time to time since midnight
* Similarly, I transformed CRASH DATE data to days since Jan 1 2020 to get a numerical value for it.
* I selected columns with numerical data
* Used the covariance matrix function to display the table of covariances



* The selected columns for df_numerical_data include key variables for analyzing accidents and their impacts. MIN PAST MIDNIGHT and DAYS SINCE 01/01/2020 provide temporal context, allowing for the examination of trends over time and time-of-day effects on accident frequency. LATITUDE and LONGITUDE are essential for mapping accident locations, revealing spatial patterns and hotspots. The columns indicating the NUMBER OF PERSONS INJURED, NUMBER OF PERSONS KILLED, and specific counts for PEDESTRIANS, CYCLISTS, and MOTORISTS affected are crucial for understanding the severity of incidents. Collectively, these variables enable a comprehensive analysis of accidents.

```python
df["MIN PAST MIDNIGHT"] = df["CRASH TIME"].apply(lambda x: (x.hour*60)+(x.minute)+(x.second/60))
start_date = pd.Timestamp('2020-01-01')
df["DAYS SINCE 01/01/2020"] = df["CRASH DATE"].apply(lambda x: x-start_date)
df_numerical_data = df[["MIN PAST MIDNIGHT", 'DAYS SINCE 01/01/2020', 'LATITUDE', 'LONGITUDE', 'NUMBER OF PERSONS INJURED', 'NUMBER OF PERSONS KILLED', 'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF PEDESTRIANS KILLED',
       'NUMBER OF CYCLIST INJURED', 'NUMBER OF CYCLIST KILLED',
       'NUMBER OF MOTORIST INJURED', 'NUMBER OF MOTORIST KILLED']]
df_numerical_data.cov()
```
